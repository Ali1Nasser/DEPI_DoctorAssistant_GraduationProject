```python
import pandas as pd

def extract_unique_records(xlsx_path, csv_path):
    # Load the Excel and CSV files
    xlsx_df = pd.read_excel(xlsx_path)
    csv_df = pd.read_csv(csv_path)
    
    # Find IDs from xlsx not in csv
    unique_ids = set(xlsx_df['ID']) - set(csv_df['ID'])
    
    # Extract records with unique IDs
    new_df = xlsx_df[xlsx_df['ID'].isin(unique_ids)]
    
    return new_df
```
