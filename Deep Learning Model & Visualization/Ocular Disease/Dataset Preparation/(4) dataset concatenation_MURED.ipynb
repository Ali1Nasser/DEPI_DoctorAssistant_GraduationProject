```python
import pandas as pd

# Load the datasets
MURED = pd.read_csv('MURED/all_data.csv')
odir_rfmid_vietai = pd.read_csv('odir_rfmid_vietai.csv')

# Display the first few rows of each dataframe for examination
MURED.head(), odir_rfmid_vietai.head()
```

```python
MURED['ID'] = range(1, len(MURED) + 1)
# Increment IDs in vietai_advance dataset based on the last ID in the full_df4_diabetic dataset
last_id = odir_rfmid_vietai['ID'].max()
MURED['ID'] = MURED['ID'].apply(lambda x: x + last_id)
MURED['Patient Age'] = 'UnKnown'
MURED['Patient Sex'] = 'UnKnown'
MURED['Left-Fundus'] = None
MURED['Right-Fundus'] = None
MURED['Left-Diagnostic Keywords'] = ''
MURED['Right-Diagnostic Keywords'] = ''
MURED['N'] = 0
MURED['D'] = 0
MURED['G'] = 0
MURED['C'] = 0
MURED['A'] = 0
MURED['H'] = 0
MURED['M'] = 0
MURED['O'] = 0
MURED['filepath'] = 0
MURED['labels'] = 0
MURED['target'] = 0
# Generate filenames
MURED['filename'] = MURED['ID'].astype(str) + '_right.jpg'
MURED['filepath'] = 'Training Images/' + MURED['filename']

# Set initial values for new columns in RFMiD dataset to merge into full_df4_diabetic

MURED['Right-Fundus'] = MURED['filename']
#rfmid_labels['Right-Fundus'] = None

# For simplicity in demonstration, create another column for right eye images (assuming all images are left for now)
MURED['filename_left'] = MURED['ID'].astype(str) + '_left.jpg'
MURED['Left-Fundus'] = MURED['filename_left']
MURED = MURED.drop(columns=['filename_left'])


MURED.head()
```

```python
def handle_disease_labels(row):
    labels = []
    diseases = ['N', 'D', 'G', 'C', 'A', 'H', 'M', 'O']
    keywords = {
        'N': 'normal',
        'D': 'diabetic',
        'G': 'glaucoma',
        'C': 'cataract',
        'A': 'macular degeneration',
        'H': 'Hypertensive',
        'M': 'myopia',
        'O': 'other'
    }
    complex_conditions = ['MH', 'TSLN', 'BRVO', 'CRVO', 'CNV', 'RS', 'ODE', 'LS', 'CSR',
                          'ASR', 'CRS', 'OTHER']

    # Determine the side (left or right) from the filename
    side = 'Left' if 'left' in row['filename'] else 'Right'

    # Check each disease and update accordingly
    if row['NORMAL'] == 1:
        row['N'] = 1
        labels.append('N')
        #row[f'{side}-Diagnostic Keywords'].append(keywords['N'])
        if row[f'{side}-Diagnostic Keywords']:
                row[f'{side}-Diagnostic Keywords'] += ', '
        row[f'{side}-Diagnostic Keywords'] += keywords['N']
    if row['DR'] == 1 or row['DN'] == 1:
        row['D'] = 1
        labels.append('D')
        #row[f'{side}-Diagnostic Keywords'].append(keywords['D'])
        if row[f'{side}-Diagnostic Keywords']:
                row[f'{side}-Diagnostic Keywords'] += ', '
        row[f'{side}-Diagnostic Keywords'] += keywords['D']
    if row['ODC'] == 1 or row['ODP'] == 1:
        row['G'] = 1
        labels.append('G')
        #row[f'{side}-Diagnostic Keywords'].append(keywords['G'])
        if row[f'{side}-Diagnostic Keywords']:
                row[f'{side}-Diagnostic Keywords'] += ', '
        row[f'{side}-Diagnostic Keywords'] += keywords['G']
    if row['ARMD'] == 1:
        row['A'] = 1
        labels.append('A')
        #row[f'{side}-Diagnostic Keywords'].append(keywords['A'])
        if row[f'{side}-Diagnostic Keywords']:
                row[f'{side}-Diagnostic Keywords'] += ', '
        row[f'{side}-Diagnostic Keywords'] += keywords['A']
    if row['HTR'] == 1:
        row['H'] = 1
        labels.append('H')
        #row[f'{side}-Diagnostic Keywords'].append(keywords['H'])
        if row[f'{side}-Diagnostic Keywords']:
                row[f'{side}-Diagnostic Keywords'] += ', '
        row[f'{side}-Diagnostic Keywords'] += keywords['H']
    if row['MYA'] == 1:
        row['M'] = 1
        labels.append('M')
        #row[f'{side}-Diagnostic Keywords'].append(keywords['M'])
        if row[f'{side}-Diagnostic Keywords']:
                row[f'{side}-Diagnostic Keywords'] += ', '
        row[f'{side}-Diagnostic Keywords'] += keywords['M']
    if any(row[cond] == 1 for cond in complex_conditions):
        row['O'] = 1
        labels.append('O')
        #row[f'{side}-Diagnostic Keywords'].append(keywords['O'])
        if row[f'{side}-Diagnostic Keywords']:
                row[f'{side}-Diagnostic Keywords'] += ', '
        row[f'{side}-Diagnostic Keywords'] += keywords['O']
    # Update labels and target columns
    row['labels'] = ', '.join(labels)
    row['target'] = [row[disease] for disease in diseases]

    return row
```

```python
MURED = MURED.apply(handle_disease_labels, axis=1)
```

```python
import os
import shutil
# Paths to the original folders and the new folder
training_path = 'MURED/images'
new_images_path = 'MURED/images2'

# Create the new folder if it doesn't exist
if not os.path.exists(new_images_path):
    os.makedirs(new_images_path)

def move_and_rename_images(df, original_folder, new_folder):
    for index, row in df.iterrows():
        # Attempt to find the file with both .png and .tif extensions
        original_filename_png = f"{row['Filename']}.png"
        original_filename_tif = f"{row['Filename']}.tif"
        original_file_path_png = os.path.join(original_folder, original_filename_png)
        original_file_path_tif = os.path.join(original_folder, original_filename_tif)
        
        if os.path.exists(original_file_path_png):
            new_filename = f"{row['filename']}"
            new_file_path = os.path.join(new_folder, new_filename)
            shutil.copy(original_file_path_png, new_file_path)
        elif os.path.exists(original_file_path_tif):
            new_filename = f"{row['filename']}"
            new_file_path = os.path.join(new_folder, new_filename)
            shutil.copy(original_file_path_tif, new_file_path)
        else:
            print(f"File not found: {original_file_path_png} or {original_file_path_tif}")
```

```python
move_and_rename_images(MURED, training_path, new_images_path)
```
