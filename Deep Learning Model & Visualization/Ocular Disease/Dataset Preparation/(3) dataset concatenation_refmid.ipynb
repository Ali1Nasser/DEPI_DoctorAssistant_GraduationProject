```python
def handle_disease_labels(row):
    labels = []
    diseases = ['N', 'D', 'G', 'A', 'H', 'M', 'O']
    keywords = {
        'N': 'normal',
        'D': 'diabetic',
        'G': 'glaucoma',
        'A': 'macular degeneration',
        'H': 'Hypertensive',
        'M': 'myopia',
        'O': 'other'
    }
    complex_conditions = ['MH', 'DN', 'BRVO', 'TSLN', 'ERM', 'LS', 'MS', 'CSR', 'CRVO', 'TV', 'ODP', 'ODE', 'ST', 'AION', 'PT', 'RT', 'RS', 'CRS', 'EDN', 'RPEC', 'MHL', 'RP', 'CWS', 'CB', 'ODPM', 'PRH', 'MNF', 'HR', 'CRAO', 'TD', 'CME', 'CF', 'VH', 'MCA', 'VS', 'BRAO', 'PLQ', 'CL']

    # Determine the side (left or right) from the filename
    side = 'Left' if 'left' in row['filename'] else 'Right'

    # Check each disease and update accordingly
    if row['Disease_Risk'] == 0:
        row['N'] = 1
        labels.append('N')
        row[f'{side}-Diagnostic Keywords'] = keywords['N']
    
    if row['DR'] == 1 or row['PTCR'] == 1:
        row['D'] = 1
        labels.append('D')
        row[f'{side}-Diagnostic Keywords'] = keywords['D']

    if row['ODC'] == 1:
        row['G'] = 1
        labels.append('G')
        row[f'{side}-Diagnostic Keywords'] = keywords['G']

    if row['ARMD'] == 1:
        row['A'] = 1
        labels.append('A')
        row[f'{side}-Diagnostic Keywords'] = keywords['A']

    if row['AH'] == 1 or row['HPED'] == 1:
        row['H'] = 1
        labels.append('H')
        row[f'{side}-Diagnostic Keywords'] = keywords['H']

    if row['MYA'] == 1:
        row['M'] = 1
        labels.append('M')
        row[f'{side}-Diagnostic Keywords'] = keywords['M']

    if any(row[cond] == 1 for cond in complex_conditions):
        row['O'] = 1
        labels.append('O')
        row[f'{side}-Diagnostic Keywords'] = keywords['O']

    # Update labels and target columns
    row['labels'] = ', '.join(labels)
    row['target'] = [row[disease] for disease in diseases]

    return row
```

```python
rfmid_labels = rfmid_labels.apply(handle_disease_labels, axis=1)
```
