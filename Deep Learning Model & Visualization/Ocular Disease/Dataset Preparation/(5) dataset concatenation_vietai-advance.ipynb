```python
import pandas as pd

# Load the datasets
vietai_advance = pd.read_csv('vietai_advance/train.csv')
odir_rfmid = pd.read_csv('odir_rfmid.csv')

# Display the first few rows of each dataframe for examination
vietai_advance.head(), odir_rfmid.head()
```

```python
vietai_advance['ID'] = range(1, len(vietai_advance) + 1)
# Increment IDs in vietai_advance dataset based on the last ID in the full_df4_diabetic dataset
last_id = odir_rfmid['ID'].max()
vietai_advance['ID'] = vietai_advance['ID'].apply(lambda x: x + last_id)
vietai_advance['Patient Age'] = 'UnKnown'
vietai_advance['Patient Sex'] = 'UnKnown'
vietai_advance['Left-Fundus'] = None
vietai_advance['Right-Fundus'] = None
#rfmid_labels['Left-Diagnostic Keywords'] = [[] for _ in range(len(rfmid_labels))]
#rfmid_labels['Right-Diagnostic Keywords'] = [[] for _ in range(len(rfmid_labels))]
vietai_advance['Left-Diagnostic Keywords'] = ''
vietai_advance['Right-Diagnostic Keywords'] = ''
vietai_advance['N'] = vietai_advance['normal']
vietai_advance['D'] = vietai_advance['diabetic retinopathy']
vietai_advance['G'] = vietai_advance['glaucoma']
vietai_advance['C'] = 0
vietai_advance['A'] = vietai_advance['macular degeneration']
vietai_advance['H'] = vietai_advance['retinal vascular occlusion']
vietai_advance['M'] = vietai_advance['opacity']
vietai_advance['O'] = vietai_advance['macular edema']
vietai_advance['filepath'] = 0
vietai_advance['labels'] = 0
vietai_advance['target'] = 0



# Generate filenames
vietai_advance['Filename'] = vietai_advance['ID'].astype(str) + '_right.jpg'
vietai_advance['filepath'] = 'Training Images/' + vietai_advance['Filename']

# Set initial values for new columns in RFMiD dataset to merge into full_df4_diabetic

vietai_advance['Right-Fundus'] = vietai_advance['Filename']
#rfmid_labels['Right-Fundus'] = None

# For simplicity in demonstration, create another column for right eye images (assuming all images are left for now)
vietai_advance['filename_left'] = vietai_advance['ID'].astype(str) + '_left.jpg'
vietai_advance['Left-Fundus'] = vietai_advance['filename_left']
vietai_advance = vietai_advance.drop(columns=['filename_left'])


vietai_advance.head()
```

```python
import os
import shutil
# Paths to the original folders and the new folder
training_path = 'vietai_advance/train'
new_images_path = 'vietai_advance/train2'

# Create the new folder if it doesn't exist
if not os.path.exists(new_images_path):
    os.makedirs(new_images_path)

def move_and_rename_images(df, original_folder, new_folder):
    for index, row in df.iterrows():
        # Attempt to find the file with both .jpg and .JPG extensions
        original_filename_jpg = f"{row['filename']}"
        original_filename_jpg_upper = f"{row['filename']}"
        original_file_path_jpg = os.path.join(original_folder, original_filename_jpg)
        original_file_path_jpg_upper = os.path.join(original_folder, original_filename_jpg_upper)
        
        if os.path.exists(original_file_path_jpg):
            new_filename = f"{row['Filename']}"
            new_file_path = os.path.join(new_folder, new_filename)
            shutil.copy(original_file_path_jpg, new_file_path)
        elif os.path.exists(original_file_path_jpg_upper):
            new_filename = f"{row['Filename']}"
            new_file_path = os.path.join(new_folder, new_filename)
            shutil.copy(original_file_path_jpg_upper, new_file_path)
        else:
            print(f"File not found: {original_file_path_jpg} or {original_file_path_jpg_upper}")
```

```python
move_and_rename_images(vietai_advance, training_path, new_images_path)
```

```python
def handle_disease_labels(row):
    labels = []
    diseases = ['N', 'D', 'G', 'C', 'A', 'H', 'M', 'O']
    keywords = {
        'N': 'normal',
        'D': 'diabetic',
        'G': 'glaucoma',
        'C': 'cataract',
        'A': 'macular degeneration',
        'H': 'Hypertensive',
        'M': 'myopia',
        'O': 'other'
    }

    # Determine the side (left or right) from the filename
    side = 'Left' if 'left' in row['filename'] else 'Right'

    # Check each disease and update accordingly
    if row['N'] == 1:
        labels.append('N')
        if row[f'{side}-Diagnostic Keywords']:
                row[f'{side}-Diagnostic Keywords'] += ', '
        row[f'{side}-Diagnostic Keywords'] += keywords['N']
    if row['D'] == 1:
        labels.append('D')
        if row[f'{side}-Diagnostic Keywords']:
                row[f'{side}-Diagnostic Keywords'] += ', '
        row[f'{side}-Diagnostic Keywords'] += keywords['D']
    if  row['G'] == 1:
        labels.append('G')
        if row[f'{side}-Diagnostic Keywords']:
                row[f'{side}-Diagnostic Keywords'] += ', '
        row[f'{side}-Diagnostic Keywords'] += keywords['G']
    if  row['C'] == 1:
        labels.append('C')
        if row[f'{side}-Diagnostic Keywords']:
                row[f'{side}-Diagnostic Keywords'] += ', '
        row[f'{side}-Diagnostic Keywords'] += keywords['C']
    if  row['A'] == 1:
        labels.append('A')
        if row[f'{side}-Diagnostic Keywords']:
                row[f'{side}-Diagnostic Keywords'] += ', '
        row[f'{side}-Diagnostic Keywords'] += keywords['A']
    if  row['H'] == 1:
        labels.append('H')
        if row[f'{side}-Diagnostic Keywords']:
                row[f'{side}-Diagnostic Keywords'] += ', '
        row[f'{side}-Diagnostic Keywords'] += keywords['H']
    if  row['M'] == 1:
        labels.append('M')
        if row[f'{side}-Diagnostic Keywords']:
                row[f'{side}-Diagnostic Keywords'] += ', '
        row[f'{side}-Diagnostic Keywords'] += keywords['M']
    if  row['O'] == 1:
        labels.append('O')
        if row[f'{side}-Diagnostic Keywords']:
                row[f'{side}-Diagnostic Keywords'] += ', '
        row[f'{side}-Diagnostic Keywords'] += keywords['O']
    
    # Update labels and target columns
    row['labels'] = ', '.join(labels)
    row['target'] = [row[disease] for disease in diseases]

    return row
```

```python
vietai_advance = vietai_advance.apply(handle_disease_labels, axis=1)
```
