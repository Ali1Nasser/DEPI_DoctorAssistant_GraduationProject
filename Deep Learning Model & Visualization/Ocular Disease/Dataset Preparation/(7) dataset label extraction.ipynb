```python
# Load and preprocess the dataset
def load_data(filepath):
    df = pd.read_csv(filepath)
    conditions = ["normal", "diabetic", "glaucoma", "cataract", "macular degeneration", "hypertensive", "myopia", "other"]
    for condition in conditions:
        df[condition] = 0

    df['Left-Diagnostic Keywords'] = df['Left-Diagnostic Keywords'].str.split('，')
    df['Right-Diagnostic Keywords'] = df['Right-Diagnostic Keywords'].str.split('，')

    def update_condition_flags(row, direction):
        keywords = row[direction + 'Diagnostic Keywords']
        for condition in conditions:
            if any(condition in keyword for keyword in keywords):
                row[condition] = 1
        if not any(row[cond] for cond in conditions): 
            row['other'] = 1   
            
        return row

    # Filter for files ending with '_right.jpg' and apply the function with 'Right-' prefix
    df_right = df[df["filename"].str.endswith("_right.jpg")].apply(lambda row: update_condition_flags(row, "Right-"), axis=1)
    # Filter for files ending with '_left.jpg' and apply the function with 'Left-' prefix
    df_left = df[df["filename"].str.endswith("_left.jpg")].apply(lambda row: update_condition_flags(row, "Left-"), axis=1)
    # Optionally, you might want to combine these two DataFrames back into one if needed
    df = pd.concat([df_right, df_left], axis=0)
    return df
```

```python
# Load data
df2 = load_data('all_datasets2.csv')
```
